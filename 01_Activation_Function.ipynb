# **激勵函數**

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras import activations
```

## ReLU (Rectified linear unit)

```python
re_x= np.linspace(-9, 9, 26)
re_x_tf = tf.constant(re_x, dtype = tf.float32)

# ReLU
re_y = activations.relu(re_x_tf).numpy()

# 繪圖
plt.plot(re_x, re_y)
plt.show()
```

## ReLU 參數

```python
re_x= np.linspace(-9, 9, 26)
re_x_tf = tf.constant(re_x, dtype = tf.float32)

# ReLU
re_y = activations.relu(re_x_tf, threshold=6).numpy()

# 繪圖
plt.plot(re_x, re_y)
plt.show()
```

```python
re_x= np.linspace(-9, 9, 26)
re_x_tf = tf.constant(re_x, dtype = tf.float32)

# ReLU
re_y = activations.relu(re_x_tf, max_value=6).numpy()

# 繪圖
plt.plot(re_x, re_y)
plt.show()
```

```python
re_x= np.linspace(-9, 9, 26)
re_x_tf = tf.constant(re_x, dtype = tf.float32)

# ReLU
re_y = activations.relu(re_x_tf, alpha=0.6).numpy()

# 繪圖
plt.plot(re_x, re_y)
plt.show()
```

## Sigmoid

```python
si_x= np.linspace(-9, 9, 26)
si_x_tf = tf.constant(si_x, dtype = tf.float32)

# Sigmoid
si_y = activations.sigmoid(si_x_tf).numpy()

# 模糊地帶
plt.axvline(-4.5, color='y')
plt.axvline(4.5, color='y')

# 繪圖
plt.plot(si_x, si_y)
plt.show()
```

## Tanh

```python
ta_x= np.linspace(-9, 9, 26)
ta_x_tf = tf.constant(ta_x, dtype = tf.float32)

# Tanh
ta_y = activations.sigmoid(ta_x_tf).numpy()

# 模糊地帶
plt.axvline(-3.5, color='y')
plt.axvline(3.5, color='y')

# 繪圖
plt.plot(ta_x, ta_y)
plt.show()
```
